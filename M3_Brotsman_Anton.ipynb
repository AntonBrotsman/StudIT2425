#### Подключаем необходимые библиотеки

```python
import csv
import numpy as np
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

#### 1.Загружаем train_data.csv

```python
TD = open(r'C:\Users\student.WIN\Downloads\Telegram Desktop\training_datasets\train_data.csv', 'r', newline='')
tr_data = pd.read_csv(TD)
```

#### 1.1 Очистим датасет и избавимся от дубликатов

```python
tr_data=tr_data.dropna()
tr_data=tr_data.drop_duplicates()
```

#### 1.2 Разбиваем датасет на обучающую и валидационную выборки

```python
n=round(0.9*len(tr_data))
train_data, valid_data = tr_data[:n], tr_data[n:] #Разбиваем датасет на выборки в отношении 9:1
```

#### 1.3 Обучаем алгоритмы машинного обучения

##### 1) Линейный дискриминантный анализ

```python
LDA=LinearDiscriminantAnalysis()
#разделяем набор на характеристики и классы
X = train_data.iloc[:, 0:3].values #характеристики
y = train_data.iloc[:, 3].values #классы
X_v = valid_data.iloc[:, 0:3].values
y_v = valid_data.iloc[:, 3].values
X_train=LDA.fit_transform(X, y)
X_valid = LDA.transform(X_v)
classifier = RandomForestClassifier(max_depth=20, random_state=23)
classifier.fit(X_train, y) #Обучаем
```

#### 1.4 Выводим точность на основе валидационной выборки

```python
y_pred = classifier.predict(X_valid)
print('Точность : ' + str(accuracy_score(y_v, y_pred)))
```
