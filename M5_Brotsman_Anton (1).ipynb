```python
from tkinter import *
import tkinter.messagebox
from tkinter import filedialog as fd
import csv
import numpy as np
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import webbrowser
```

#### Реализуем модель

```python
LDA=LinearDiscriminantAnalysis()

def upl_traind(path):
    TD = open(path, 'r', newline='')
    tr_data = pd.read_csv(TD)
    return tr_data

def processing_data(tr_data):
    tr_data=tr_data.dropna()
    tr_data=tr_data.drop_duplicates()
    return tr_data

def separation(tr_data):
    n=round(0.9*len(tr_data))
    train_data, valid_data = tr_data[:n], tr_data[n:] #Разбиваем датасет на выборки в отношении 9:1
    return train_data, valid_data

def sep_for_classes(train_data, valid_data):
    #разделяем набор на характеристики и классы
    X = train_data.iloc[:, 0:3].values #характеристики
    y = train_data.iloc[:, 3].values #классы
    X_v = valid_data.iloc[:, 0:3].values
    y_v = valid_data.iloc[:, 3].values
    return X, y, X_v, y_v

def training(X, y):
    X_train=LDA.fit_transform(X, y)
    classifier = RandomForestClassifier(max_depth=20, random_state=23)
    classifier.fit(X_train, y) #Обучаем
    return classifier

def check_accurancy(X_v, y_v, classifier):
    X_valid = LDA.transform(X_v)
    y_pred = classifier.predict(X_valid)
    print('Точность : ' + str(accuracy_score(y_v, y_pred)))

def prediction(test_path, classifier, out_path):
    TESTD = open(test_path, 'r', newline='')
    ts_data = pd.read_csv(TESTD)
    X_test = LDA.transform(ts_data.values)
    y_pred = classifier.predict(X_test)
    df = pd.DataFrame(y_pred)
    df.to_csv(out_path, sep=',', index=False, encoding='utf-8')
```

```python
def LDA_model(path, test_path, out_path):
    tr_data = upl_traind(path)
    tr_data = processing_data(tr_data)
    train_data, valid_data = separation(tr_data)
    X, y, X_v, y_v = separation_for_classes(train_data, valid_data)
    classifier = training(X, y)
    check_accurancy(X_v, y_v, classifier)
    prediction(test_path, classifier, out_path)
```
